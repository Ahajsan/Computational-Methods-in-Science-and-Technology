```python
import numpy as np
import matplotlib.pyplot as plt
from bitalg.visualizer.main import Visualizer

def zero_function(x):
    return x ** 2 - 25 * np.sin(x) ** 10
    
def zero_aprx():
    return 0.754774

def ro():
    return [1e-8, 1e-12, 1e-16]
```

```python
def next_step(a,b,f):
    return b - (f(b) * (b - a))/(f(b) - f(a))

def print_siecznych_metoda(a,b):
    vis = Visualizer()
    f = zero_function
    xs = np.linspace(0, 2.5,1000)
    ys = f(xs)
    points = []
    
    for i in range(len(ys)):
        point = (xs[i], ys[i])
        points.append(point)
    vis.add_point(points)
    point_a = vis.add_point((a,f(a)),color = 'red')
    point_b = vis.add_point((b,f(b)),color = 'red')
    line = vis.add_line(((a,f(a)),(b,f(b))), color = 'yellow')
    iteration = 0;
    while True:    
        iteration += 1
        x = next_step(a,b,f)
        point_x = vis.add_point((x,f(x)), color='orange')
        if abs(f(x)) < 1e-8: 
            vis.save_gif(filename='siecz_res_10-8_[1;2]', interval=300)
            return x
        if f(x) * f(a) < 0:
            vis.remove_figure(point_b)
            b = x
            point_b = vis.add_point((b,f(b)),color = 'red')
            vis.remove_figure(point_x)
            vis.remove_figure(line)
            line = vis.add_line(((a,f(a)),(b,f(b))), color = 'yellow')
        else:
            vis.remove_figure(point_a)
            a = x
            point_a = vis.add_point((a,f(a)),color = 'red')
            vis.remove_figure(point_x)
            vis.remove_figure(line)
            line = vis.add_line(((a,f(a)),(b,f(b))), color = 'yellow')
        if iteration > 1000:
            vis.save_gif(filename='siecz_res_10-8_[1;2]', interval=300)
            return x
        


print(print_siecznych_metoda(1,2))
        
    
    
```

```python
def next_step(a,b,f):
    return b - (f(b) * (b - a))/(f(b) - f(a))

def print_siecznych_metoda(a,b,N,e,method):
    start = (a,b)
    max_iterations = N
    f = zero_function
    
    if f(a) * f(b) == 0:
        return a if f(a) == 0 else b
 
    x = next_step(a, b, f)
    iteration = 1
    prev_x = x

    while iteration < max_iterations:

        
        if method == 1 and abs(f(x)) < e:
            return f"Przedział: {start}, Wynik: {x}, Iteracje: {iteration-1}, Różnica: {abs(zero_aprx() - x):}"

        if f(x) * f(a) < 0:
            b = x
        else:
            a = x

        new_x = next_step(a, b, f)
        if method == 2 and abs(new_x - x) < e:
            return f"Przedział: {start}, Wynik: {x}, Iteracje: {iteration-1}, Różnica: {abs(zero_aprx() - x):}"
            
        x = new_x
        iteration += 1
    return "maksymalna ilosć iteracji osiaągnięta", f"Przedział: {start}, Wynik: {x}, Różnica: {abs(zero_aprx() - x):}"
        

r = ro()
for e in r:
    a = 0.4
    b = 2
    print(e)
    while True:
        print(print_siecznych_metoda(a,b,1000,e,1))
        b -= 0.1
        if b < a:
            print()
            break
    
```

```python
def zero_derative_function(x):
    return 2*x - 250 * np.cos(x) * np.sin(x) ** 9

def next_point(x, f, d, max_iterations, iteration):
    if iteration >= max_iterations:
        return None  
    return x - f(x) / d(x)  
    
def Newton(x, N, r, method):
    start = x
    max_iterations = N
    iteration = 0
    f = zero_function
    d = zero_derative_function
    n = next_point
    
    while iteration < max_iterations:
        iteration += 1
        if method == 1 and abs(f(x)) < r:
            return f"Przedział: {start}, Wynik: {x}, Iteracje: {iteration}, Różnica: {abs(zero_aprx() - x)}"
        
        next_x = next_point(x, f, d, max_iterations, iteration)
        if next_x is None:
            return "maksymalna ilość iteracji osiągnięta", f"Przedział: {start}, Wynik: {x}, Różnica: {abs(zero_aprx() - x)}"
        
        if method == 2 and abs(f(x) - f(next_x)) < r:
            return f"Przedział: {start}, Wynik: {x}, Iteracje: {iteration}, Różnica: {abs(zero_aprx() - x)}"
        
        x = next_x
    
    return "maksymalna ilość iteracji osiągnięta", f"Przedział: {start}, Wynik: {x}, Różnica: {abs(zero_aprx() - x)}"



r = ro()
for e in r:
    print(e)
    z = 0.4
    while True:
        print(Newton(z,1000,e,1))
        z += 0.1
        if z > 2.1:
            print()
            break
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicja funkcji
def f(x):
    return x ** 2 - 25 * np.sin(x) ** 10

# Przedział i siatka punktów
x = np.linspace(0.4, 2.0, 1000)
y = f(x)

# Rysowanie wykresu
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = x^2 - 25 \cdot \sin^{10}(x)$', color='blue')
plt.axhline(0, color='gray', linestyle='--')
plt.title("sd")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
