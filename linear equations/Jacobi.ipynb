```python
import numpy as np
import random
import time
```

```python
real_x1 = np.array([1 if random.randint(1, 10) % 2 == 0 else -1 for _ in range(1000)])
```

```python
def jacobi1(A, b, r, z, max_iter=1000):
    n = len(b)
    x = np.zeros(n)
    x_new = np.zeros(n)

    start_time = time.perf_counter()

    for k in range(max_iter):
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]

        if z == 1 and np.linalg.norm(x_new - x) < r:
            end_time = time.perf_counter()
            total_time = end_time - start_time
            return x_new, k + 1, total_time / (k + 1)

        elif z == 2 and np.linalg.norm(A @ x_new - b) < r:
            end_time = time.perf_counter()
            total_time = end_time - start_time
            return x_new, k + 1, total_time / (k + 1)

        x = x_new.copy()

    end_time = time.perf_counter()
    total_time = end_time - start_time
    return x, max_iter, total_time / max_iter

def spectral_radius_jacobi(A):
    D = np.diag(np.diag(A))
    L_plus_U = A - D
    D_inv = np.diag(1 / np.diag(A))
    M = -D_inv @ L_plus_U
    eigenvalues = np.linalg.eigvals(M)
    return max(abs(eigenvalues))


def pre_jacobi1(real_x, n, r, z):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if j == i:
                A[i, j] = 9
            elif j > i:
                A[i, j] = (-1)**j * 2 / j
            elif j == i - 1:
                A[i, j] = 2 / i
            else:
                A[i, j] = 0

    b = A @ real_x
    solution, iterations, avg_time = jacobi1(A, b, r, z)
    error = np.max(abs(solution - real_x))

    return f"n: {n}, błąd maksymlany: {error}, iteracje: {iterations}, średni czas/iterację: {avg_time:.6e} s"
```

```python
def jacobi2(A, b, r, z, max_iter=1000):
    n = len(b)
    x = np.random.normal(loc=0, scale=1, size=n)
    x_new = x.copy()

    start_time = time.perf_counter()

    for k in range(max_iter):
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]

        if z == 1 and np.linalg.norm(x_new - x) < r:
            end_time = time.perf_counter()
            total_time = end_time - start_time
            return x_new, k + 1, total_time / (k + 1)

        elif z == 2 and np.linalg.norm(A @ x_new - b) < r:
            end_time = time.perf_counter()
            total_time = end_time - start_time
            return x_new, k + 1, total_time / (k + 1)

        x = x_new.copy()

    end_time = time.perf_counter()
    total_time = end_time - start_time
    return x, max_iter, total_time / max_iter

def pre_jacobi2(real_x, n, r, z):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if j == i:
                A[i, j] = 9
            elif j > i:
                A[i, j] = (-1)**j * 2 / j
            elif j == i - 1:
                A[i, j] = 2 / i
            else:
                A[i, j] = 0

    b = A @ real_x
    solution, iterations, avg_time = jacobi2(A, b, r, z)
    error = np.max(abs(solution - real_x))

    return f"n: {n}, błąd maksymlany: {error}, iteracje: {iterations}, średni czas/iterację: {avg_time:.6e} s"
```
